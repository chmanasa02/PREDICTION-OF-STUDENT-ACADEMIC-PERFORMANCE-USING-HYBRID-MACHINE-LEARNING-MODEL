```python
# load and summarize the dataset
from pandas import read_csv
from collections import Counter
from matplotlib import pyplot
from sklearn.preprocessing import LabelEncoder
data1 = data.values
# split into input and output elements
X, y = data1[:, :-1], data1[:, -1]
# label encode the target variable
y = LabelEncoder().fit_transform(y)
# summarize distribution
counter = Counter(y)
for k,v in counter.items():
	per = v / len(y) * 100
	print('Class=%d, n=%d (%.3f%%)' % (k, v, per))
# plot the distribution
pyplot.bar(counter.keys(), counter.values())
pyplot.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('Target',axis=1), data['Target'], test_size=0.3,random_state=0)
```

```python
#SMOTE
from imblearn.over_sampling import SMOTE
X_smote,y_smote=SMOTE(k_neighbors=3).fit_resample(X_train,y_train)
y_smote.value_counts().plot(kind='bar')
```

```python
y_smote = LabelEncoder().fit_transform(y_smote)
```

```python
print(y_smote)
```

```python
y_test = LabelEncoder().fit_transform(y_test)
```

```python
#K-Nearst Neighbor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
skfold=StratifiedKFold(n_splits=10)
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(X_smote,y_smote)
prediction1 = knn.predict(X_test)
acc_KNN=accuracy_score(y_test,prediction1)*100
print("Accuracy of KNN: ",acc_KNN )#only smote
print("F1 score: ", f1_score(y_test,prediction1,pos_label='positive',average='micro'))
print("Recall: ", recall_score(y_test,prediction1,pos_label='positive',average='micro'))
print("Precision: ", precision_score(y_test,prediction1,pos_label='positive',average='micro'))
scores=cross_val_score(knn,X_smote,y_smote,cv=skfold)
print(scores)
print(np.mean(scores))#smote+skfold




#AUC CURVE
from yellowbrick.classifier import ROCAUC
visualizer= ROCAUC(knn.fit(X_smote,y_smote),classes=["Dropout","Graduate"])
visualizer.fit(X_smote,y_smote)
visualizer.score(X_test,y_test)
visualizer.show()
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(max_iter=10000)
logmodel.fit(X_smote,y_smote)
LogisticRegression(max_iter=10000)
prediction2 = logmodel.predict(X_test)
acc_LR=accuracy_score(y_test,prediction2)*100
print("Accuracy of LR: ", acc_LR)
print("F1 score: ", f1_score(y_test,prediction2,pos_label='positive',average='micro'))
print("Recall: ", recall_score(y_test,prediction2,pos_label='positive',average='micro'))
print("Precision: ", precision_score(y_test,prediction2,pos_label='positive',average='micro'))

scores=cross_val_score(logmodel,X_smote,y_smote,cv=skfold)
print(scores)
print(np.mean(scores))


from yellowbrick.classifier import ROCAUC
visualizer= ROCAUC(logmodel.fit(X_smote,y_smote),classes=["Dropout","Graduate"])
visualizer.fit(X_smote,y_smote)
visualizer.score(X_test,y_test)
visualizer.show()
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_smote,y_smote)
predictions = dtree.predict(X_test)
acc_DT=accuracy_score(y_test,predictions)*100
print("Accuracy of DT: ", acc_DT)
print("F1 score: ", f1_score(y_test,predictions,pos_label='positive',average='micro'))
print("Recall: ", recall_score(y_test,predictions,pos_label='positive',average='micro'))
print("Precision: ", precision_score(y_test,predictions,pos_label='positive',average='micro'))
print("\n")
#print(classification_report(y_test,predictions))
print("\n")
#print(confusion_matrix(y_test,predictions))

scores=cross_val_score(dtree,X_smote,y_smote,cv=skfold)
print(scores)
print(np.mean(scores))


#AUC Curve
from yellowbrick.classifier import ROCAUC
visualizer= ROCAUC(dtree.fit(X_smote,y_smote),classes=["Dropout","Graduate"])
visualizer.fit(X_smote,y_smote)
visualizer.score(X_test,y_test)
visualizer.show()
```

```python
# Define estimators of all
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
estimator_list = [
    ('knn',knn),
    ('logmodel',logmodel),
    ('dtree',dtree),
    ('rfc',rfc),
    ('model',model),
    ('gnb',gnb),
    ('mlp',mlp),
    ('xgboost',xgboost),
    ('adaboost',adaboost) ]
# Build stack model
stack_model = StackingClassifier(
    estimators=estimator_list, final_estimator=LogisticRegression(max_iter=3000),cv=10)
# Train stacked model
stack_model.fit(X_smote, y_smote)
# Make predictions
y_train_pred = stack_model.predict(X_smote)
y_test_pred = stack_model.predict(X_test)
# Training set model performance
stack_model_train_accuracy = accuracy_score(y_smote, y_train_pred) # Calculate Accuracy
stack_model_test_accuracy = accuracy_score(y_test, y_test_pred) # Calculate Accuracy
print('Model performance for Training set')
print('- Accuracy: %s' % stack_model_train_accuracy)
print('----------------------------------')
print('Model performance for Test set')
print('- Accuracy: %s' % stack_model_test_accuracy)


#AUC
from yellowbrick.classifier import ROCAUC
visualizer= ROCAUC(stack_model,classes=["Dropout","Graduate"])
visualizer.fit(X_smote,y_smote)
visualizer.score(X_test,y_test)
visualizer.show()


y_pred = stack_model.predict_proba(X_test)[:,1]
auc = metrics.roc_auc_score(y_test, y_pred)
print(f"AUC: {auc:.3f}")
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from itertools import combinations
from sklearn.ensemble import VotingClassifier
models = [
    ('lr', LogisticRegression(random_state=42)),
    ('knn', KNeighborsClassifier(n_neighbors=5)),
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('svm', SVC(kernel='linear', random_state=42)),
    ('nb', GaussianNB()),
    ('nn', MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=1000, random_state=42)),
    ('adaboost', AdaBoostClassifier()),
    ('xgboost', xgb.XGBClassifier())

]

# Generate all possible combinations of the models
combinations_list = []
for i in range(1, len(models) + 1):
    combinations_list += list(combinations(models, i))

# Evaluate the accuracy of each combination
for combination in combinations_list:
    voting_clf = VotingClassifier(estimators=list(combination), voting='hard')
    voting_clf.fit(X_smote, y_smote)
    y_pred = voting_clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy of combination {}: {:.2f}%".format([x[0] for x in combination], accuracy * 100))
```
